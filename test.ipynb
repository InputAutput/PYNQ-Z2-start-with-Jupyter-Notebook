```python
from pynq.overlays.base import BaseOverlay
import time

base = BaseOverlay("base.bit")

leds = base.leds                # LD0..LD3
switches = base.switches        # SW0..SW3

try:
    while True:
        if switches[0].read():
            leds[0].on()        # SW0=1 LD0=1
        else:
            leds[0].off()       # SW0=0 LD0=0
        time.sleep(0.02)
except KeyboardInterrupt:
    leds[0].off()
```
